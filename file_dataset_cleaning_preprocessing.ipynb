```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', None, 'David'], 'Age': [25, None, 30, 40]}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Name'].fillna('Unknown', inplace=True)

print("\nCleaned DataFrame:\n", df)
```
